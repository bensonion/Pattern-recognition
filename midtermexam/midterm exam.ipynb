```python
import sympy as sm
x,y= sm.symbols('x,y')
f=exp(-(((x-1)**2)+((y+1)**2)))
fx= sm.diff(f,x)
fy= sm.diff(f,y)
a= sm.Eq(fx, 0) 
b= sm.Eq(fy, 0) 
sm.solve([a, b],x,y,  
         dict=True,  
        )
```
